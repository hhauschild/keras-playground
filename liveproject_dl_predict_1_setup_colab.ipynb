```
# build model with keras
from keras.models import Sequential 
from keras.layers import Activation, Dense, LSTM
from keras import losses 
from keras import optimizers 
from keras import metrics 

model = Sequential() 
the_layer = Dense(1, activation = 'linear', input_dim = 1) 
model.add(the_layer) 

model.compile(loss='mean_squared_error', optimizer='adam')
```

```
# training data. F=C∗1.8+32.
y_train=np.random.randint(0,30,size=(20)) # celsius
x_train=y_train*1.8+32 # farenheit
print([x_train,y_train])
# validation data
y_val=np.random.randint(0,30,size=(20)) # celsius
x_val=y_val*1.8+32 # farenheit
```

```
# train model

model.fit(x_train, y_train, batch_size = 5, epochs = 1000, validation_data = (x_val, y_val))
```
